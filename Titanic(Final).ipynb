```python
import numpy as np
import pandas as pd
```

```python
#Loading the Train Dataset 
train_data=pd.read_csv('train.csv')
test_data=pd.read_csv('test.csv')
```

```python
train_data.head()
```

```python
train_data.isnull().sum()
```

```python
#Creating a copy of the dataset
data=train_data
```

```python
# Droping the unneccseary Columns from the dataset
data=train_data.drop(columns=["Name","Ticket","Cabin"])
```

```python
data
```

```python
#Appling Feature Scalling to column "Sex"
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
data["Sex"]=le.fit_transform(data["Sex"])
```

```python
data
```

```python
data.info()
```

```python
data["Embarked"].unique()
```

```python
# As there are only 3 null values in the Embarked so  we will delete those rows
data.dropna(subset=["Embarked"],inplace=True)
```

```python
data.info()
```

```python
# Replacing all the null values of Age with mean 
data["Age"].fillna(data["Age"].mean(),inplace=True)
```

```python
data.info()
```

```python
##Appling Feature Scalling to column "Embarked"

data["Embarked"]=le.fit_transform(data["Embarked"])
```

```python
data
```

```python
y=data.iloc[:,1].values
```

```python
data_new=data.drop(columns=["Survived","PassengerId"])
```

```python
data_new
```

```python
X=data_new.iloc[:,0:].values
```

```python
X
```

```python
from sklearn.model_selection import train_test_split 
from sklearn import tree
from sklearn.metrics import accuracy_score
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=1)
```

```python
clf = tree.DecisionTreeClassifier()
clf = clf.fit(X_train, y_train)
```

```python

prediction=clf.predict(X_test)
prediction
```

```python
print("Accuracy is-",accuracy_score(y_test,prediction)*100)
```
